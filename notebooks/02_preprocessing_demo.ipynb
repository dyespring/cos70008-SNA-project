# 02 - Preprocessing Demo

Demonstrate the text cleaning and tokenisation pipeline for both policy and Yelp data sources.

```python
import sys
sys.path.insert(0, '..')

from src.ingest.pdf_reader import read_policy_pdf, pages_to_document
from src.ingest.yelp_reader import load_yelp_reviews
from src.preprocessing.cleaner import clean_text
from src.preprocessing.tokeniser import SpacyTokeniser
```

## 1. Raw vs Cleaned Text (Policy)

```python
pages = read_policy_pdf()
raw_sample = pages[2][1]  # a content page
print('--- RAW ---')
print(raw_sample[:500])

cleaned = clean_text(raw_sample, source='policy')
print('\n--- CLEANED ---')
print(cleaned[:500])
```

## 2. Tokenisation and Lemmatisation

```python
tokeniser = SpacyTokeniser()
doc = tokeniser.process(cleaned, doc_id='sample_page', source='policy')

print(f'Sentences: {len(doc.sentences)}')
print(f'Total lemmas (after stop-word removal): {len(doc.all_lemmas)}')

# Show first 3 sentences
for i, sent in enumerate(doc.sentences[:3]):
    print(f'\nSentence {i}: {sent.text[:100]}...')
    print(f'  Lemmas: {sent.lemmas[:10]}')
    print(f'  Entities: {sent.entities}')
    print(f'  Noun chunks: {sent.noun_chunks[:5]}')
```

## 3. Vocabulary Analysis

```python
from collections import Counter

lemma_freq = Counter(doc.all_lemmas)
print(f'Unique lemmas: {len(lemma_freq)}')
print('\nTop 20 lemmas:')
for word, count in lemma_freq.most_common(20):
    print(f'  {word}: {count}')
```

## 4. Raw vs Cleaned Text (Yelp)

Compare the same pipeline on Yelp review data.

```python
df = load_yelp_reviews(category_filter='Restaurants', sample_n=5)

raw_review = df.iloc[0]['text']
print('--- RAW YELP REVIEW ---')
print(raw_review[:500])

cleaned_review = clean_text(raw_review, source='yelp')
print('\n--- CLEANED ---')
print(cleaned_review[:500])

yelp_doc = tokeniser.process(cleaned_review, doc_id='review_0', source='yelp')
print(f'\nSentences: {len(yelp_doc.sentences)}')
print(f'Lemmas: {len(yelp_doc.all_lemmas)}')
print(f'Source: {yelp_doc.source}')

for i, sent in enumerate(yelp_doc.sentences[:3]):
    print(f'\nSentence {i}: {sent.text[:100]}...')
    print(f'  Lemmas: {sent.lemmas[:10]}')
    print(f'  Entities: {sent.entities}')
    print(f'  Noun chunks: {sent.noun_chunks[:5]}')
```
