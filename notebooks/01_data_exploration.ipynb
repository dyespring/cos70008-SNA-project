## 1. Policy Document (PDF)

```python
from src.ingest.pdf_reader import read_policy_pdf, pages_to_document

pages = read_policy_pdf()
print(f'Total pages extracted: {len(pages)}')
print(f'Page numbers: {[p for p, _ in pages]}')

# Preview first page
page_num, text = pages[0]
print(f'\n--- Page {page_num} (first 500 chars) ---')
print(text[:500])
```

```python
full_text = pages_to_document(pages)
words = full_text.split()
print(f'Total words: {len(words):,}')
print(f'Total characters: {len(full_text):,}')
print(f'Unique words: {len(set(w.lower() for w in words)):,}')
```

## 2. Yelp Reviews (JSON)

The Yelp dataset is pre-extracted into `data/Yelp JSON/yelp_dataset/`. We stream-read and sample.

```python
from src.ingest.yelp_reader import load_yelp_reviews

print(f'Yelp data directory: {YELP_EXTRACTED_DIR}')

# Load a small sample of restaurant reviews
df = load_yelp_reviews(category_filter='Restaurants', sample_n=100)
print(f'Reviews loaded: {len(df)}')
print(f'Columns: {list(df.columns)}')
df.head()
```
